# FEATURE ENGINEERING

```python
import pandas as pd
import numpy as np

# Assuming 'data' is your original DataFrame

# Reset the index to avoid KeyError
data.reset_index(drop=True, inplace=True)

# Example: Extract titles from the 'Name' column
data['Title'] = data['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

# Example: Create a new feature 'FamilySize' by combining 'SibSp' and 'Parch'
data['FamilySize'] = data['SibSp'] + data['Parch'] + 1

# Example: Create a binary 'IsAlone' feature based on 'FamilySize'
data['IsAlone'] = np.where(data['FamilySize'] == 1, 1, 0)

# Example: Convert 'Sex' to a numerical feature
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})

# Example: Binning 'Age' into age groups
bins = [0, 18, 35, 50, np.inf]
labels = ['0-18', '19-35', '36-50', '51+']
data['AgeGroup'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# Example: One-hot encode categorical variables (e.g., 'Embarked' and 'Pclass')
data = pd.get_dummies(data, columns=['Embarked', 'Pclass'], prefix=['Embarked', 'Pclass'])

# Display the modified DataFrame
print(data.head())
```
